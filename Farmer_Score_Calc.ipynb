```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/CreditScoreData - Sheet1.csv" , parse_dates= True)
```

```python
df
```

```python

daily_income = df.groupby('Date' , sort = False)['I/E'].sum().reset_index()


daily_income['daily_status'] = daily_income['I/E'].apply(lambda x: 1 if x > 0 else -1)

df_new = pd.DataFrame(daily_income)
```

```python
df_new
```

```python
weekly_sum = df_new['daily_status'].rolling(window=7).sum()
weekly_mode = weekly_sum.apply(lambda x: -1 if x < 0 else 1)

```

```python
df_new['weekly_mode'] = weekly_mode
```

```python
df_new
```

```python
weekly_sum = df_new['weekly_mode'].rolling(window=7).sum()
```

```python
def calculate_score(weekly_modes):
      score = 250
      consecutive_count = 0
      for mode in weekly_modes:
        if mode == 1:
            score += 1
            consecutive_count = max(0, consecutive_count + 1)
        elif mode == -1:
            score -= 1
            consecutive_count = max(0, consecutive_count + 1)
        else:
            consecutive_count = 0

        if consecutive_count == 3:
            if mode == 1:
                score += 5
            elif mode == -1:
                score -= 5

      return score



```

```python
cf_sc  = calculate_score(df_new['weekly_mode'])
```
